```python
%matplotlib inline

import pandas as pd
import ipywidgets as widgets
import numpy as np

from matplotlib import pyplot as plt
from pathlib import Path

data_path = Path('data', 'clean')
```

```python
# loading the data
data_file = Path(data_path, 'net_fiscal_balance.csv')
balance = pd.read_csv(data_file)
```

```python
# create an array with the delta value between row_index_x and row_index_y

def delta(df, row_index_x, row_index_y, label='Difference'):
    d = [label]
    for i in range(1, len(df.columns)):
        d.append(abs(df.iloc[row_index_x, i] - df.iloc[row_index_y, i]))
    
    return d
```

## Net Fiscal Balance in Value (Â£ million)

```python
# only the revenue in value
values = balance[0:3].copy().reset_index(drop=True)

# swaping indexes
values_t = values.set_index('Period').transpose()


plt.rcParams['figure.figsize'] = (10,5)

f = plt.figure()
plt.title('Net Fiscal Balance (Â£ million)', color='black')
values_t.plot(ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
plt.show()
```

### Excluding North Sea

```python
# only the revenue in value
values_ens = balance.take([0,2]).copy().reset_index(drop=True)

# And add a delta between them
values_ens.loc[len(values_ens.index)] = delta(values_ens, 0, 1, 'Difference (Â£ million)')

# swaping indexes
values_ens_t = values_ens.set_index('Period').transpose()


plt.rcParams['figure.figsize'] = (10,5)

f = plt.figure()
plt.title('Net Fiscal Balance (Â£ million)', color='black')
values_ens_t.plot(ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
plt.show()
```

```python
# only the revenue in value
values_ins = balance.take([1,2]).copy().reset_index(drop=True)

# And add a delta between them
values_ins.loc[len(values_ins.index)] = delta(values_ins, 0, 1, 'Difference (Â£ million)')

# swaping indexes
values_ins_t = values_ins.set_index('Period').transpose()

plt.rcParams['figure.figsize'] = (10,5)

f = plt.figure()
plt.title('Net Fiscal Balance (Â£ million)', color='black')
values_ins_t.plot(ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
plt.show()
```
