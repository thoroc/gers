```python
%matplotlib inline

import pandas as pd
import ipywidgets as widgets
import numpy as np

from matplotlib import pyplot as plt
from pathlib import Path

data_path = Path('data', 'clean')
```

```python
# loading the data
data_file = Path(data_path, 'current_budget_balance.csv')
budget = pd.read_csv(data_file)
```

```python
# create an array with the delta value between row_index_x and row_index_y

def delta(df, row_index_x, row_index_y, label='Difference'):
    d = [label]
    for i in range(1, len(df.columns)):
        d.append(abs(df.iloc[row_index_x, i] - df.iloc[row_index_y, i]))
    
    return d
```

## Current Budget Balance in Value (£ million)

```python
# only the revenue in value
values = budget[0:2].copy().reset_index(drop=True)
# swaping indexes
values_t = values.set_index('Period').transpose()

plt.rcParams['figure.figsize'] = (10,5)

f = plt.figure()
plt.title('Current Budget Balance in Value (£ million)', color='black')
values_t.plot(ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
plt.show()
```

## Current Budget Balance in % of UK Revenue

```python
# only the revenue in Percent of the UK revenue
percent = budget[2:4].copy().reset_index(drop=True)
percent_t = percent.set_index('Period').transpose()

plt.rcParams['figure.figsize'] = (10,5)

f = plt.figure()
plt.title('Current Budget Balance (% of UK Revenue)', color='black')
percent_t.plot(ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
plt.show()
```

## Current Budget Balance in % of GDP

```python
gdp = budget[4:].copy().reset_index(drop=True)
gdp_t = gdp.set_index('Period').transpose()

plt.rcParams['figure.figsize'] = (10,5)

f = plt.figure()
plt.title('Current Budget Balance (% GDP)', color='black')
gdp_t.plot(ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
plt.show()
```
